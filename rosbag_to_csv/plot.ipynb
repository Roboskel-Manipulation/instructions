```python
import numpy as np
import pandas as pd
import plotly
import plotly.offline as py
import plotly.graph_objects as go

py.init_notebook_mode(connected=True)

file = pd.read_csv("points.csv")

fig = go.Figure(data=go.Scatter(x=file['/topic/x'], y=file['/topic/y'], mode='markers'),
               layout=go.Layout(title='Example Plot', xaxis=dict(title='x(m)'), yaxis=dict(title='y(m)')))

py.iplot(fig)
```
